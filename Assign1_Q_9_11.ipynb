9. Scatter Plot Visualization:
‚óè Create a scatter plot using both Excel/Python to visually inspect the correlation between x and y.


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [10, 20, 30, 40, 50]
y = [5, 10, 15, 20, 25]
```

```python
plt.scatter(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Scatter Plot of x vs y")
# Display the plot
plt.grid(True)
plt.show()
```

10. Box Plot Visualization:
‚óè Create a box plot for the dataset to visualize Q1, Q2, Q3, lower bound, upper bound, and outliers.

```python
sorted_datax=sorted(x)
print("Sorted Data:", x)
q1x = np.percentile(sorted_datax, 25)
q2x = np.percentile(sorted_datax, 50)
q3x = np.percentile(sorted_datax, 75)
iqrx = q3x - q1x
lower_boundx= q1x - 1.5 * iqrx
upper_boundx= q3x + 1.5 * iqrx
outliersx = []
for value in x:
    if value < lower_boundx or value > upper_boundx:
        outliersx.append(value)

sorted_datay=sorted(y)
print("Sorted Data:", y)
q1y = np.percentile(sorted_datay, 25)
q2y = np.percentile(sorted_datay, 50)
q3y = np.percentile(sorted_datay, 75)
iqry = q3y - q1y
lower_boundy= q1y - 1.5 * iqry
upper_boundy= q3y + 1.5 * iqry
outliersy = []
for value in x:
    if value < lower_boundy or value > upper_boundy:
        outliersy.append(value)
```

```python
# Create box plots for x and y
plt.figure(figsize=(10, 3))  # Adjust figure size for better view

plt.subplot(1, 2, 1)
plt.boxplot(x, tick_labels=['x'])  # Use tick_labels instead of labels
plt.title('Box Plot of x')
plt.ylabel('Values')

plt.subplot(1, 2, 2)
plt.boxplot(y, tick_labels=['y'])  # Use tick_labels instead of labels
plt.title('Box Plot of y')
plt.ylabel('Values')

plt.tight_layout()
plt.show()
```

11. Histogram Analysis:
‚óè Construct a histogram to show the frequency distribution of the dataset.



```python
# Create histograms for x and y
plt.figure(figsize=(10, 3))

plt.subplot(1, 2, 1)
plt.hist(x, bins=5, edgecolor='black')
plt.title('Histogram of Dataset x')
plt.xlabel('Values')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(y, bins=5, edgecolor='black')
plt.title('Histogram of Dataset y')
plt.xlabel('Values')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
